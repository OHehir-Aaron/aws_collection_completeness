```python
import matplotlib.pyplot as plt
import matplotlib
import numpy as np
import seaborn as sns

def read_file_to_list(filename):
    """
    Reads all lines from a file into a list using file.readlines().

    Args:
        filename (str): The path to the file.

    Returns:
        list: A list of strings, where each string is a line from the file (including newline characters).
    """
    try:
        with open(filename, 'r') as file:
            # readlines() returns a list where each element is one line
            lines_list = file.readlines()
            out = []
            for i in lines_list:
                i = i.replace('\n', '')
                out.append(float(i))
        return out
    except FileNotFoundError:
        print(f"Error: The file '{filename}' was not found.")
        return []
    except IOError as e:
        print(f"An I/O error occurred: {e}")
        return []

landsat9_2024 = read_file_to_list('landsat9_L2-L2_aws_2024.txt')
landsat8_2024 = read_file_to_list('landsat8_L2-L2_aws_2024.txt')
landsat9_2025 = read_file_to_list('landsat9_L2-L2_aws_2025.txt')
landsat8_2025 = read_file_to_list('landsat8_L2-L2_aws_2025.txt')

sentinel2a_2024 = read_file_to_list('sentinel2a_L2-L2_aws_2024.txt')
sentinel2a_2025 = read_file_to_list('sentinel2a_L2-L2_aws_2025.txt')
sentinel2b_2024 = read_file_to_list('sentinel2b_L2-L2_aws_2024.txt')
sentinel2b_2025 = read_file_to_list('sentinel2b_L2-L2_aws_2025.txt')
sentinel2c_2024 = read_file_to_list('sentinel2c_L2-L2_aws_2024.txt')
sentinel2c_2025 = read_file_to_list('sentinel2c_L2-L2_aws_2025.txt')

wofs_2025 = read_file_to_list('wofs_L2-L3_aws_2025.txt')
fc_2025 = read_file_to_list('fc_L2-L3_aws_2025.txt')

landsat9_2024_data = np.array(landsat9_2024)
landsat8_2024_data = np.array(landsat8_2024)
landsat9_2025_data = np.array(landsat9_2025)
landsat8_2025_data = np.array(landsat8_2025)

sentinel2a_2024_data = np.array(sentinel2a_2024)
sentinel2a_2025_data = np.array(sentinel2a_2025)
sentinel2b_2024_data = np.array(sentinel2b_2024)
sentinel2b_2025_data = np.array(sentinel2b_2025)
sentinel2c_2024_data = np.array(sentinel2c_2024)
sentinel2c_2025_data = np.array(sentinel2c_2025)

wofs_2025_data = np.array(wofs_2025)
fc_2025_data = np.array(fc_2025)

plt.hist([landsat9_2024_data, landsat9_2025_data], bins=10, alpha=.6, label=['landsat 9 2024 Uploads', 'landsat 9 2025 Uploads'], color=['blue', 'orange'])

# Add labels, a title, and a legend for clarity
plt.xlabel('Percentage of level 2s upload to AWS (per region code)')
plt.ylabel('Count')
plt.title('Landsat 9 Completeness - Uploads to s3 (per region code) as %')
plt.legend(loc='upper left')

# Display the plot
plt.show()

```
